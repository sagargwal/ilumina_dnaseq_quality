```python
# sample data "A3_R1.fastq"
import random
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import numpy as np
def betterqc(file_path):
    with open(f"{file_path}", "r") as file:
        data = file.read()
    dnaArr = data.split('\n')
    uniqueId = []
    uniqueIdIndex = 0
    sequence = []
    sequenceIndex = 1
    score = []
    scoreIndex = 3
    i = 0
    while uniqueIdIndex < len(dnaArr):
        uniqueId.append(dnaArr[uniqueIdIndex])
        uniqueIdIndex += 4
        i += 1
    i = 0
    while sequenceIndex < len(dnaArr):
        sequence.append(dnaArr[sequenceIndex])
        sequenceIndex += 4
        i += 1
    i = 0
    while scoreIndex < len(dnaArr):
        score.append(dnaArr[scoreIndex])
        scoreIndex += 4
        i += 1
#     print(len(uniqueId))
#     print(len(sequence))
#     print(len(score))
    dat =  {"sequence":sequence,"score":score}
    data =pd.DataFrame(dat)
    def fred(char1):
        ascii_value1 = ord(char1)
        return str(ascii_value1 - 33)
    def fredconv(score):
        result= []
        for i in score:
            d =[]
            for j in i:
                d.append(fred(j))
            result.append(",".join(d))
        return result
    free = fredconv(score)
    fred_score= {"fred_score":fredconv(score)}
    data = pd.concat([data,pd.DataFrame(fred_score)],axis=1)
    g= []
    for i in data.fred_score:
        aaa = i.split(",")
        d = len(aaa)
        g.append(d)
    maax = min(g)
    arrr =[]
    l= 0 
    for i in range(maax):
        fot = []
        for j in free:
            s = j.split(",")
            fot.append(int(s[l]))
        arrr.append(fot)
        l+=1
    fig = plt.figure(figsize =(17,15))
    ax = fig.add_axes([1,1,1,1])
    np.random.seed(12)
    first_ten = np.arange(10)
    random_numbers = np.random.choice([i for i in range(maax) if i not in first_ten], 50, replace=False)
    indices_to_plot = np.concatenate((first_ten, np.sort(random_numbers)))
    selected_x = [arrr[i] for i in indices_to_plot if i<maax]
    means = [np.mean(k) for k in selected_x]
    count =[]
    for f in range(1,len(indices_to_plot)+1):
        count.append(f)
    
    plt.plot(count, means, marker='o', linestyle='-', color='r') 
    plt.boxplot(selected_x)
    custom_ticks = indices_to_plot
    da =[]
    for n in range(1,len(indices_to_plot)+1):
        da.append(n)
    plt.xticks(da,indices_to_plot)
    plt.xlabel("Position in read (bp)",fontsize=17)
    plt.ylabel("Phred score",fontsize = 17)
    plt.title("Quality in different positions",fontsize = 20)
    return plt.show()

```

```python
def quality_distribution(file_path):
    with open(f"{file_path}", "r") as file:
        data = file.read()
    dnafile_path = data.split('\n')
    uniqueId = []
    uniqueIdIndex = 0
    sequence = []
    sequenceIndex = 1
    score = []
    scoreIndex = 3
    i = 0
    while uniqueIdIndex < len(dnafile_path):
        uniqueId.append(dnafile_path[uniqueIdIndex])
        uniqueIdIndex += 4
        i += 1
    i = 0
    while sequenceIndex < len(dnafile_path):
        sequence.append(dnafile_path[sequenceIndex])
        sequenceIndex += 4
        i += 1
    i = 0
    while scoreIndex < len(dnafile_path):
        score.append(dnafile_path[scoreIndex])
        scoreIndex += 4
        i += 1
#     print(len(uniqueId))
#     print(len(sequence))
#     print(len(score))
    dat =  {"sequence":sequence,"score":score}
    data =pd.DataFrame(dat)
    def fred(char1):
        ascii_value1 = ord(char1)
        return str(ascii_value1 - 33)
    def fredconv(score):
        result= []
        for i in score:
            d =[]
            for j in i:
                d.append(fred(j))
            result.append(",".join(d))
        return result
    free = fredconv(score)
    fred_score= {"fred_score":fredconv(score)}
    data = pd.concat([data,pd.DataFrame(fred_score)],axis=1)
    g= []
    for i in data.fred_score:
        aaa = i.split(",")
        d = len(aaa)
        g.append(d)
    maax = min(g)
    arrr =[]
    l= 0 
    for i in range(maax):
        fot = []
        for j in free:
            s = j.split(",")
            fot.append(int(s[l]))
        arrr.append(fot)
        l+=1
    mean_quality=[]
    for h in arrr:
        meen = np.mean(h)
        mean_quality.append(round(meen))
    unii = np.unique(mean_quality)
    unii_count =[]
    for i in range(len(unii)):
        c= 0
        for j in mean_quality:
            if j == unii[i]:
                c+=1
        unii_count.append(c)
    plt.figure(figsize= (12,7))
    plt.plot(unii,unii_count,marker='o', linestyle='-')
    plt.xticks([i for i in range(min(unii),(max(unii)+1))])
    plt.xlabel("Mean Sequence Quality(phred score)")
    plt.title("Quality Distribution")
        
```
